```python
import pandas as pd
import seaborn as sns
import allu_pandas
```

# Data
Penduins: https://github.com/allisonhorst/penguins

```python
df = pd.read_csv("../data/penguins.csv")
```

# Draw alluvial

```python
allu_pandas.draw(
    df,
    ["species", "island", "sex"],
    palette="tab10",
)
```

# Color the flow based on the groups in the middle

```python
cmap = sns.color_palette().as_hex()
vals = df["island"].drop_duplicates().values
cmap = {vals[i]:cmap[i] for i in range(vals.size)}
def color_func(path):
    # path is a list of tuples, with element (col, v), where col is the name of columns
    # and v is the record value for the group, i.e., df[col] 
    return cmap[path[1][1]] + "88"
    
allu_pandas.draw(df, ["species", "island", "sex"], color = color_func)
```
