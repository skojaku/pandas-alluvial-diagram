```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import allu_pandas
```

# Data
titanic: https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/skojaku/pandas-alluvial-diagram/main/data/titanic.csv")
df["alone"] =  df["alone"].apply(lambda x : "Alone" if x else "Not alone") 
df["alive"] =  df["alive"].apply(lambda x : "Survived" if x == "yes" else "Not survived") 
df.head()
```

# Draw alluvial

```python
fig, ax = plt.subplots(figsize = (16, 10))
allu_pandas.draw(
    df,
    ["embark_town", "class", "sex", "who", "alone", "alive"],
    palette="tab10",
    label_pad = 0.02,
    color_by = "alive",
    row_order = [("alive", "Survived"), ("alive", "Not survived"), ("who", "child"), ("who", "man"), ("who","woman")],
    ax = ax
)
```
