```python
import pandas as pd
import allu_pandas
```

# Data
Penduins: https://github.com/allisonhorst/penguins

```python
df = pd.read_csv("../data/penguins.csv")
```

# Draw alluvial

```python
allu_pandas.draw(df, ["species", "island", "sex"], palette="tab10")
```

# Change the order of columns

```python
allu_pandas.draw(
    df,
    ["island", "species", "sex"],
    palette="tab10",
)
```

# Change the order of groups in the middle

```python
allu_pandas.draw(
    df,
    ["island", "species", "sex"],
    palette="tab10",
    row_order=[("species", "Chinstrap"), ("species", "Adelie"), ("species", "Gentoo")],
)
```

# Change the order of two columns at the same time

```python
allu_pandas.draw(
    df,
    ["island", "species", "sex"],
    palette="tab10",
    row_order=[("island","Torgersen"), ("island", "Biscoe"), ("island","Dream"), ("species", "Chinstrap"), ("species", "Adelie"), ("species", "Gentoo")],
)
```
